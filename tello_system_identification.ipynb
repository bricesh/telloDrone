```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy
```

```python
filename = 'logs/blackbox-20_Jan_2019_10_49_37.log'
```

```python
log_df = pd.read_csv(filename, index_col=False) ## add sep=';'
```

```python
log_df.iloc[297:462, :]
```

```python
print('x-axis std:', np.mean([np.std(log_df.iloc[158:200, :].target_x), np.std(log_df.iloc[382:462, :].target_x), np.std(log_df.iloc[1304:1337, :].target_x)]))
```

```python
print('y-axis std:', np.mean([np.std(log_df.iloc[158:200, :].target_y), np.std(log_df.iloc[382:462, :].target_y), np.std(log_df.iloc[1304:1337, :].target_y)]))
```

```python
print('z-axis std:', np.mean([np.std(log_df.iloc[158:200, :].target_area), np.std(log_df.iloc[382:462, :].target_area)]))
```
